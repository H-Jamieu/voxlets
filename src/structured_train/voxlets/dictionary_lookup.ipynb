```python
'''
idea is to load a depth image, and for each pixel work out the dictionary item which belongs to it
the ultimate aim is to be able to create a training set based on the dictionary
'''
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt 
import cPickle as pickle
import sys
sys.path.append('/Users/Michael/projects/shape_sharing/src/structured_train/')

from thickness import paths
from thickness import voxel_data
from thickness import mesh
from thickness import images
```

```python
shoebox_kmeans_path = paths.base_path + "voxlets/shoebox_dictionary_training.pkl"
km = pickle.load(open(shoebox_kmeans_path, 'rb'))
print km.cluster_centers_.shape
```

```python
# load in an image, a mesh, a voxel grid etc.
modelname = paths.modelnames[10]
view = paths.views[25]

im = images.CroppedRGBD()
im.load_bigbird_from_mat(modelname, view)
im.disp_channels()

cam = mesh.Camera()
cam.load_bigbird_matrices(modelname, view)
cam.adjust_intrinsic_scale(0.5) # as the image is half-sized

im.set_camera(cam)
# projecting the depth points from the camera out into world space
world_xyz = im.get_world_xyz()
world_normals = im.get_world_normals()
```

```python
# loading the voxel grid for this object
vgrid = voxel_data.BigBirdVoxels()
vgrid.load_bigbird(modelname)
```

```python
# now find the shoebox for each pixel
shoebox_p_from_origin = np.array((0.1, 0.05, 0.1))
shoebox_voxelsize = 0.2/20.0

pix_to_use = np.array(np.nonzero(~np.isnan(im.frontrender).flatten()))[0]

print pix_to_use.shape
print world_xyz.shape

model_shoeboxes = []
points = []

for count, pix_idx in enumerate(pix_to_use):
    point = world_xyz[pix_idx]
    points.append(point)
    
    norm = world_normals[pix_idx]
    
    shoebox = voxel_data.ShoeBox((20, 20, 20))
    shoebox.set_p_from_grid_origin(shoebox_p_from_origin)
    shoebox.set_voxel_size(shoebox_voxelsize)
    shoebox.initialise_from_point_and_normal(point, norm, np.array([0, 0, 1]))
    
    shoebox_xyz_in_world = shoebox.world_meshgrid()
    shoebox_xyx_in_world_idx, valid = vgrid.world_to_idx(shoebox_xyz_in_world, True)
    
    # fill in the shoebox voxels
    idxs = shoebox_xyx_in_world_idx[valid, :]
    occupied_values = vgrid.extract_from_indices(idxs)
    shoebox.set_indicated_voxels(valid, occupied_values)
    
    model_shoeboxes.append(shoebox)
        
    if np.mod(count, 1000) == 0:
        print "Done " + str(count) + ' of ' + str(pix_to_use.shape)

#uv = cam.project_points(np.array(points))
#plt.imshow(im.rgb)
#plt.hold(True)
#plt.plot(uv[:, 0] - im.aabb[0], uv[:, 1]- im.aabb[2], '.')
#plt.hold(False)
```

```python
# for each shoebox, find the closest item in the dictionary
all_boxes = np.array([box.V.flatten().astype(np.float16) for box in model_shoeboxes])
cluster_idxs = km.predict(all_boxes)
```

```python
# now recreate an image showing these cluster idxs
plt.rcParams['figure.figsize'] = (10.0, 8.0)

cluster_idx_im = np.copy(0*im.mask.flatten())
cluster_idx_im[pix_to_use] = cluster_idxs
cluster_idx_im = cluster_idx_im.reshape(im.mask.shape)

print cluster_idx_im.shape

plt.subplot(121)
plt.imshow(im.rgb)
plt.subplot(122)
plt.imshow(cluster_idx_im, interpolation='Nearest')
```
