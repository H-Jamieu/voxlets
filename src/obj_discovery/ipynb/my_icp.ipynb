```python
labels_dir = '/media/ssd/data/oisin_house/obj_discovery/labelling/'

# load labels and create inverse mapping
scene_to_labels = yaml.load(open(labels_dir + 'labels.yaml'))

labels_to_scene = {}

for scene, labels in scene_to_labels.items():
    for idx, label in labels.items():
        if label == '':
            continue
        elif label in labels_to_scene:
            labels_to_scene[label].append((scene, idx))
        else:
            labels_to_scene[label] = [(scene, idx)]


# create mapping from a scene name to all the sequences of this scene
scene_to_sequence = {}

for scene in paths.all_train_data:
    if scene['scene'] in scene_to_sequence:
        scene_to_sequence[scene['scene']].append(scene)
    else:
        scene_to_sequence[scene['scene']] = [scene]

            
```

```python
def crop_image_to_mask(img, mask):
    '''returns a cropped version of the image tightly cropped on the mask'''
    left = np.argmax(mask.sum(0) > 0)
    top = np.argmax(mask.sum(1) > 0)
    right = mask.shape[1] - np.argmax(mask.sum(0)[::-1] > 0)
    bottom = mask.shape[0] - np.argmax(mask.sum(1)[::-1] > 0)
    if len(img.shape) == 3:
        return img[top:bottom, left:right, :]
    else:
        return img[top:bottom, left:right]
    
    
def floor_plane_to_3d(labels_floor_plane, sc):
    '''expands a grid to 3D. Code modified from scenes.py'''
    labels3d = np.expand_dims(labels_floor_plane, axis=2)
    labels3d = np.tile(labels3d, (1, 1, sc.gt_tsdf.V.shape[2]))

    labels_3d_grid = sc.gt_tsdf.copy()
    labels_3d_grid.V = labels3d
    labels_3d_grid.V[:, :, :sc.floor_height] = 0

    return labels_3d_grid


def load_images_of_object(obj_name):
    '''returns a list of rgb images of the named object'''
    imgs = []
    xyzs = []
    ims = []
    
    scene_idxs = labels_to_scene[obj_name]
    for scene_name, idx in scene_idxs:

        if scene_name not in scene_to_sequence:
            print "Skipping", scene_name
            continue

        sequences = scene_to_sequence[scene_name]

        for sequence in sequences:
            # load the sequence
            sc = scene.Scene(0.025)
            sc.load_sequence(sequence, frame_nos=0, segment_with_gt=False,
                segment=False, save_grids=False, carve=False)
            
            # load the mask
            labels_floor_plane = \
                scipy.io.loadmat(labels_dir + 'labels/' + sequence['scene'] + '.mat')['seg']

            labels_3d = floor_plane_to_3d(labels_floor_plane, sc)
            labels_im = sc.im.label_from_grid(labels_3d)
            mask = labels_im==idx
            
            temp = crop_image_to_mask(sc.im.depth, mask)
            xyz = sc.im.get_world_xyz()[mask.ravel(), :]
            
            imgs.append(temp)
            xyzs.append(xyz)
            ims.append(sc.im)

    return imgs, xyzs, ims

imgs, xyzs, ims = load_images_of_object('bunny')
```

```python
# ok now form an error metric using the reprojected points
def calc_error(fixed_im, floating_cloud, R, t):
    # transform pc1 into the world space of pc0
    floating_in_fixed_im_world = \
        icp._transform(floating_cloud, R, t)

    # project this into camera 1
    floating_uvd = fixed_im.cam.project_points(floating_in_fixed_im_world)
    
    # now assess the quality of the match
    # first get all the fixed depths at the positions the floating projects to:
    uv = np.round(floating_uvd[:, :2]).astype(int)
    fixed_d = fixed_im.depth[uv[:, 0], uv[:, 1]]

    # now form the distances between each pair...
    diffs = fixed_d - floating_uvd[:, 2]
    diffs = diffs[~np.isnan(diffs)]
    
    # now extract the three types of diffs
    thresh = 0.02
    close_diffs = diffs[np.abs(diffs) < thresh]
    floating_in_front = diffs[diffs > thresh]
    floating_behind = diffs[diffs < -thresh]
    
    # put these together
    energy = 0
    # close_diffs are good, they will contribute the 
    energy += np.sum(close_diffs**2) 
#     / close_diffs.shape[0]
    
    #     floating_in_front are very bad, and should be penalised quite a lot - maybe a fixed price per one
    energy += floating_in_front.shape[0] * (0.15**2)
    
    #     floating_behind are a bit bad, should be penalised just a bit
    energy += floating_behind.shape[0] * (0.001**2)
    
    plt.hist(diffs, 100)
    return energy
    
    

# this can then be used to work out the way it should be done
for R, t, error, inliers in possible_transforms:

    # transform pc1 into the world space of pc0
    xyz_0_in_pc0_world_space = icp._transform(xyz_for_icp[1], R, t)

    # project this into camera 1
    cam1 = ims[icp_imgs[1]].cam
    xyz_0_in_cam_1 = cam1.project_points(xyz_0_in_pc0_world_space)

    # get the original points in cam1
    xyz_1_in_cam_1 = cam1.project_points(xyz_for_icp[0])

    fixed_im = ims[icp_imgs[0]]
    floating_cloud = xyz_for_icp[1]
    print calc_error(fixed_im, floating_cloud, R, t)
```
