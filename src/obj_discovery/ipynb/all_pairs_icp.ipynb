```python
def get_min_var_idx(ims, masks):
    """Returns the idx of the region with the smallest variance"""

    def get_obj_xyz_var(im, mask):
        """returns the variance of the xyz from an image"""
        xyz = im.get_world_xyz()[mask.ravel()]
        return np.linalg.det(np.cov(xyz.T))

    variances = []
    for i in range(len(ims)):

        im = ims[i]
        mask = masks[i]
        variances.append(get_obj_xyz_var(im, mask))

    variances = np.array(variances)
    med = np.median(variances)
    return np.argmin(np.abs(variances-med))

start_idx = get_min_var_idx(ims, masks)
```
