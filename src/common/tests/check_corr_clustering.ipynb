```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import sys, os
sys.path.append(os.path.expanduser('~/projects/shape_sharing/src/'))
from common import corr_clust
import numpy as np
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
iris = sklearn.datasets.load_iris()
X = iris['data']
Y = iris['target']

corr = corr_clust.CorrClust(
    forest_params={'n_estimators':100},
    balance_training=True)
corr.train(X, Y, 500)
Y_pred, energy = corr.test(X)
```

```python
# plt.plot(Y, Y_pred[0], 'o')
conf_mat = confusion_matrix(Y, Y_pred)
plt.imshow(conf_mat, interpolation='nearest', cmap=plt.get_cmap('gray_r'))
plt.colorbar()
```

```python
from sklearn import metrics
metrics.adjusted_rand_score(Y, Y_pred)
```

```python
# checking edges
temp_X, temp_Y = corr._form_pairs(X, Y)
pred_edges = corr.rf.predict_proba(temp_X)[:, 1]

t0, e0 = np.histogram(pred_edges[temp_Y==0], bins=np.linspace(0, 1, 10))
t1, e1 = np.histogram(pred_edges[temp_Y==1], bins=np.linspace(0, 1, 10))

plt.plot(e0[:-1], t0, 'r')
plt.plot(e1[:-1], t1, 'b')
```

```python
# viewing the weights matrix
plt.imshow(corr.weights, interpolation='nearest', cmap=plt.get_cmap('bwr'))
plt.colorbar()
```
